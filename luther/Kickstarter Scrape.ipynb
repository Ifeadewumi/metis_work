```python
# if needed: conda install requests
import requests
from bs4 import BeautifulSoup
import re
import time
import datetime
from fake_useragent import UserAgent

ua = UserAgent()
```

```python
def return_urls_from_page(page_url):
    user_agent = {'User-agent': ua.random}
    response  = requests.get(page_url, headers = user_agent)
    project_urls_regex = re.compile('https://www.kickstarter.com/projects/[\d|a-z|\-]*/[\d|a-z|\-]*')
    all_projects_urls = re.findall(project_urls_regex, response.text)
    return list(set(all_projects_urls))

def get_all_urls(url_prefix, topic='None', page_start = 1, page_end = 2):

    current_time = str(datetime.datetime.now())
    data = open(topic + '_' + current_time + '_{}_{}.txt'.format(page_start, page_end), "a" ) #open file in append mode
#     data.write('New run from page {}\n'.format(page_start)) #separator in file
    all_urls = []
    page_numbers = list(range(page_start, page_end + 1))
    for i in page_numbers:
        url_to_scrape = url_prefix.format(str(i))
        urls_in_page = return_urls_from_page(url_to_scrape)
        data.write(('\n').join(urls_in_page)) # append new data
        data.write('\n')
        if (i % 10 == 0):
            print('Finished Page {}'.format(str(i)))
            data.write('Finish Page {}\n'.format(str(i)))
        all_urls.extend(urls_in_page)
        time.sleep(1)
    data.close()
    return all_urls
```
