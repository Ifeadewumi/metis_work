```python
import numpy as np
import pandas as pd
from sklearn.utils import resample
```

```python
df = pd.read_csv('./pair_boot.csv')
```

```python
df.head()
```

```python
df.shape[0]
```

```python
med = []
for i in np.arange(1, 200, 1):
    temp = resample(df).y
    med.append(temp.median())
```

```python
import statsmodels.stats.api as sms

sms.DescrStatsW(med).tconfint_mean()
```

```python
import numpy as np, scipy.stats as st

st.t.interval(0.95, len(med)-1, loc=np.mean(med), scale=st.sem(med))
```

```python
from sklearn.linear_model import LinearRegression
```

```python
betas = []
for i in np.arange(0,200,1):
    lr = LinearRegression()
    resample_df = resample(df)
    X = resample_df.drop('y', axis=1)
    y = resample_df.y
    lr.fit(X, y)
    print(np.append(lr.coef_, lr.intercept_))
```

```python
betas_df = pd.DataFrame(betas)
```
